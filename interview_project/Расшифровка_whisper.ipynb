# Если в тексте остались мета-символы разметки \ тайм-коды

```python
import re

with open('00274.txt') as file:
  text = file.read()

def clean_text():
  tags = re.compile(r'[[0-9].*[0-9]]')
  cleaned = tags.sub('', text)
  return cleaned

cleaned = clean_text()
file_path = 'output_result00274.txt'

with open(file_path, 'w') as file:
  file.write(cleaned)

print(f'output result was saved to {file_path}')
```
